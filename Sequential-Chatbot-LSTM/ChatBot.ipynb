```python
import pandas as pd
import numpy as np

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense

from sklearn.preprocessing import LabelEncoder
```

```python
questions = [
    "Hi",
    "Hello",
    "How are you?",
    "What is your name?",
    "Who created you?",
    "What can you do?",
    "How old are you?",
    "Where are you from?",
    "What time is it?",
    "What is today's date?",
    "Tell me a joke",
    "What is AI?",
    "What is Machine Learning?",
    "What is Python?",
    "What is Java?",
    "What is HTML?",
    "What is CSS?",
    "What is JavaScript?",
    "What is the capital of India?",
    "What is the capital of the USA?",
    "Who is the Prime Minister of India?",
    "What is the largest planet?",
    "What is the fastest animal?",
    "What is 5 + 7?",
    "What is the square root of 64?",
    "Thank you",
    "Bye",
    "Good morning",
    "Good night",
    "I love you"
]

answers = [
    "Hello! How can I help you today?",
    "Hi there! Nice to meet you.",
    "I'm doing great! Thanks for asking. How about you?",
    "My name is ChatBot.",
    "I was created by a developer to assist users.",
    "I can answer questions, chat with you, and provide information.",
    "I don't have an age because I'm a computer program.",
    "I exist in the digital world!",
    "Sorry, I can't check the current time unless connected to a clock or API.",
    "I can't access today's date unless connected to a calendar API.",
    "Why don't programmers like nature? It has too many bugs!",
    "AI stands for Artificial Intelligence. It enables computers to perform tasks that usually require human intelligence.",
    "Machine Learning is a branch of AI where computers learn from data to make predictions or decisions.",
    "Python is a popular programming language known for its simplicity and versatility.",
    "Java is an object-oriented programming language widely used for web, mobile, and enterprise applications.",
    "HTML is the standard language used to create web pages.",
    "CSS is used to style and design HTML web pages.",
    "JavaScript is a programming language used to make web pages interactive.",
    "New Delhi is the capital of India.",
    "Washington, D.C. is the capital of the United States.",
    "Narendra Modi is the Prime Minister of India.",
    "Jupiter is the largest planet in our solar system.",
    "The peregrine falcon is the fastest animal when diving.",
    "5 + 7 = 12.",
    "The square root of 64 is 8.",
    "You're welcome! Happy to help.",
    "Goodbye! Have a wonderful day.",
    "Good morning! Hope you have a fantastic day.",
    "Good night! Sleep well and take care.",
    "Thank you! I'm happy to chat with you and help however I can."
]
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(answers)
print(y)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(questions)
X = tokenizer.texts_to_sequences(questions)
print(X)
```

```python
max_length = max(len(i) for i in X)
X = pad_sequences(X, maxlen=max_length, padding="post")
print(X)
```

```python
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=16))
model.add(LSTM(32))
model.add(Dense(len(np.unique(y)), activation="softmax"))
```

```python
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X, y, epochs=300)
```

```python
question = ["What is AI?"]
test = tokenizer.texts_to_sequences(question)
test = pad_sequences( test, maxlen=max_length, padding="post")
```

```python
prediction = model.predict(test)
index = np.argmax(prediction)
print(index)
```

```python
answer = encoder.inverse_transform([index])
print(answer[0])
```

```python
while True:

    question = input("You : ")

    if question.lower() == "exit":
        break

    test = tokenizer.texts_to_sequences([question])

    test = pad_sequences(
        test,
        maxlen=max_length,
        padding="post"
    )

    prediction = model.predict(test, verbose=0)

    index = np.argmax(prediction)

    answer = encoder.inverse_transform([index])

    print("Bot :", answer[0])
```

```python
model.save('Chatbot.keras')
```
