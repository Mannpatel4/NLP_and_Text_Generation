# RNN

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.utils import to_categorical
```

```python
with open("/content/Dataset_RNN.txt", "r", encoding="utf-8") as f:
    data = f.read().lower()
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
print(tokenizer.word_index)
```

```python
input_sequences = []

for line in data.split("\n"):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram = token_list[:i+1]
        input_sequences.append(n_gram)
print(input_sequences)
```

```python
max_len = max(len(seq) for seq in input_sequences)
print(max_len)
```

```python
input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
print(input_sequences)
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]
print(X.shape)
print(y.shape)
```

```python
total_words = len(tokenizer.word_index)+1
```

```python
y = to_categorical(y, num_classes=total_words)
```

```python
model = Sequential()

model.add(Embedding(total_words, 100, input_length=max_len-1))

model.add(LSTM(150, return_sequences=True))

model.add(LSTM(100))

model.add(Dense(total_words, activation="softmax"))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=200, verbose=1)
```

```python
def predict_next_word(text):

    token_list = tokenizer.texts_to_sequences([text])[0]

    token_list = pad_sequences(
        [token_list],
        maxlen=max_len-1,
        padding='pre'
    )

    predicted = np.argmax(model.predict(token_list, verbose=0))

    for word, index in tokenizer.word_index.items():
        if index == predicted:
            return word
```

```python
print(predict_next_word("What men of name resort to"))
```

```python
print(predict_next_word("At Pembroke, or at Harford-west, in"))
```

```python
print(predict_next_word("That in the sty of this most"))
```

```python
print(predict_next_word("Oxford, redoubted Pembroke, Sir"))
```
